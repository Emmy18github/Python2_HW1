````python
class Player :
    """Player class has name, a position and speed attributes."""

    def __init__(self, name, position) :
        """Initialization method with name and position parameters"""
        self.name = name
        self.position = position
        self.speed = 1
        

    def print_info(self) :            
        print(f"Player = {self.name}, position = {self.position}, speed = {self.speed}.")
````

````python
#create player objects
player1 = Player("Lamar Jackson", "QB")
player2 = Player("Devonta Freeman", "RB")

#print player info before
player1.print_info()
player2.print_info()
print("")

#update player1 speed
player1.speed = 8

#print player info after
player1.print_info()
player2.print_info()
print("")
````

## Exercise 2: 
Create a class called Rectangle with height and length attributes and has following methods: 
- perimeter() : returns a perimeter of a rectangle object
- area() : returns an area of a rectangle object

Then,
1. create two instances of Rectangle class
2. print each rectangle's perimeter and determine which rectangle has larger perimeter
3. print each rectangle's area and determine which rectangle has larger area

````python
class Rectangle:
    """Rectangle class has height and length attributes."""
    
    def __init__(self, height, length):
        self.height = height
        self.length = length

        
    def perimeter(self):
        """Returns a perimeter of a rectangle"""
        return 2 * (self.length + self.height)
    
    
    def area(self):
        """Returns an area of a rectangle"""
        return self.length * self.height
    
    
    def print_info(self, value1, value2, type):
        """Print two rectangle objects perimeter values.  Print which one is larger"...OR...
           Print two rectangle objects area values.  Print which one is larger.  
           Pass type to identify if perimeter or area.  """
        
        print(f"{type}1 = {value1} and {type}2 = {value2}")
        if value1 > value2:
            print(f"{type}1 is greater than {type}2!")
        elif value1 < value2:
            print(f"{type}2 is greater than {type}1!") 
        else:        
            print(f"{type}1 is equal to {type}2!")
        print("")
        return

````

````python
#Main Rectangle Program
rectangle1 = Rectangle(20, 10)
rectangle2 = Rectangle(40, 20)

perimeter1 = rectangle1.perimeter()
perimeter2 = rectangle2.perimeter()

rectangle1.print_info(perimeter1, perimeter2, "Perimeter")

area1 = rectangle1.area()
area2 = rectangle2.area()

rectangle1.print_info(area1, area2, "Area")
        
    
````

## Exercise 3: 
Create a class called Triangle with height and base attributes and has following methods:

- perimeter() : returns a perimeter of a right triangle object
- area() : returns an area of a right triangle object

Then,

- create two instances of Triangle class
- print each triangle's perimeter and determine which triangle has larger perimeter
- print each triangle's area and determine which triangle has larger area

Use 
- perimeter of a right triangle = height + base + Square-Root((height \** 2) + (base \** 2)) 
- area of a right triangle = ( height * base ) / 2

You'll need to <b>import math</b> library and use <b>math.sqrt()</b> function.



````python
import math
class Triangle:
    """Triangle class with height and base attributes."""
    
    def __init__(self, height, base):
        self.height = height
        self.base = base
        
    def perimeter(self):
        """Returns a perimeter of a right triangle object"""
        return self.height + self.base + math.sqrt((self.height**2) + (self.base**2)) 
    
    def area(self):
        """Returns an area of a right triangle object"""
        return (self.height * self.base) / 2
    
    
    def print_info(self, value1, value2, type):
        """Print two triangle objects perimeter values.  Print which one is larger"...OR...
           Print two triangle objects area values.  Print which one is larger.  
           Pass type to identify if perimeter or area.  """
        
        print(f"{type}1 = {value1} and {type}2 = {value2}")
        if value1 > value2:
            print(f"{type}1 is greater than {type}2!")
        elif value1 < value2:
            print(f"{type}2 is greater than {type}1!") 
        else:        
            print(f"{type}1 is equal to {type}2!")
        print("")
        return

        
````
